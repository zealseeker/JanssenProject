## Computing overlap between LINCS L1000 gene expression data and the Janssen image data set

```python
# import packages
import pandas as pd
```

```python
# read in metadata (phase 1 and 2) and Janssen compound InChI keys
meta1 = pd.read_csv("../Input/LINCS_Metadata/sig_info/GSE70138_Broad_LINCS_sig_info_2017-03-06.txt",sep="\t")
meta2 = pd.read_csv("../Input/LINCS_Metadata/sig_info/GSE92742_Broad_LINCS_sig_info.txt",sep="\t")
image_comps = pd.read_csv("../Input/janssen_compounds.csv")
```

```python
# Explore image data compounds
image_comps
```

```python
# Extract pert info (LINCS compound inchi keys)
pert_1 = pd.read_csv("../Input/LINCS_Metadata/pert_info/GSE70138_Broad_LINCS_pert_info (1).txt",sep="\t")
pert_2 = pd.read_csv("../Input/LINCS_Metadata/pert_info/GSE92742_Broad_LINCS_pert_info.txt",sep="\t")
```

```python
# Explore pert info
pert_1
```

```python
pert_2
```

```python
# Extract InChI keys from pert info
all_perts = pert_1['inchi_key'].tolist() + pert_2['inchi_key'].tolist()
```

```python
# Get unique set of InChI keys
all_perts = list(set(all_perts))
```

```python
# List all InChI keys
all_perts
```

```python
len(all_perts) #21,178 unique compounds in LINCS
```

```python
# Extract Janssen InChI keys
janssen_comps = image_comps['Compound InChIKey'].tolist()
```

```python
# List keys
janssen_comps
```

```python
# Function to compute overlap between two lists
def intersection(lst1, lst2): 
    lst3 = [value for value in lst1 if value in lst2] 
    return lst3 
overlapping_comps = intersection(lst1=janssen_comps,lst2=all_perts)
```

```python
# List overlapping compounds between LINCS and Janssen
overlapping_comps
```

```python
len(overlapping_comps) #665 overlapping compounds
```

```python
# Subset Janssen df to retain only overlapping keys
image_comps_w_lincs = image_comps[image_comps['Compound InChIKey'].isin(overlapping_comps)]
image_comps_w_lincs
```

```python
# Now subset the pert info dataframes
pert_inames1 = pert_1[pert_1['inchi_key'].isin(overlapping_comps)]
pert_inames2 = pert_2[pert_2['inchi_key'].isin(overlapping_comps)]
```

```python
pert_inames1
```

```python
pert_inames2
```

```python
# Extract pert_inames to subset metadata files
pert_inames = pert_inames1['pert_iname'].tolist() + pert_inames2['pert_iname'].tolist()
```

```python
# Get unique list of pert_inames
pert_inames = list(set(pert_inames))
pert_inames
```

```python
len(pert_inames) # 662 unique compound names
```

```python
# Now get exp info from metadata (i.e. sig_ids)
exp1 = meta1[meta1['pert_iname'].isin(pert_inames)]
exp2 = meta2[meta2['pert_iname'].isin(pert_inames)]
```

```python
exp1
```

```python
exp2
```

```python
# Only retain A549, HEPG2, WPMY, WPMY-1 cell lines
cells1 = exp1[exp1['cell_id'].isin(['A549','HEPG2','WPMY','WPMY-1'])]
```

```python
cells2 = exp2[exp2['cell_id'].isin(['A549','HEPG2','WPMY','WPMY-1'])]
```

```python
# Extract compounds in LINCS and Janssen and in the correct cell lines
cells_comps = cells1['pert_iname'].tolist() + cells2['pert_iname'].tolist()
```

```python
# Get unique inchi keys for all overlapping compounds in correct cell lines
inchi_1 = pert_inames1[pert_inames1['pert_iname'].isin(cells_comps)]
inchi_2 = pert_inames2[pert_inames2['pert_iname'].isin(cells_comps)]
inchi_all = inchi_1['inchi_key'].tolist() + inchi_2['inchi_key'].tolist()
inchi_all_u = list(set(inchi_all))
len(inchi_all_u)
```

```python
# Subset Janssen df with final set of compounds
comps_to_send = image_comps[image_comps['Compound InChIKey'].isin(inchi_all_u)]
```

```python
comps_to_send
```

```python
# Save as .csv file
comps_to_send.to_csv("../Output/lincs_compounds_overlapping.csv")
```

```python
# Now subset metadata for extraction script
# First get pert_inames
pert_inames_final_1 = pert_1[pert_1['inchi_key'].isin(inchi_all_u)]
pert_inames_final_2 = pert_2[pert_2['inchi_key'].isin(inchi_all_u)]
```

```python
pert_inames_final_1
```

```python
pert_inames_final = pert_inames_final_1['pert_iname'].tolist() + pert_inames_final_2['pert_iname'].tolist()
pert_inames_final = list(set(pert_inames_final))
pert_inames_final
```

```python
# Subset sig info file
exp1final = meta1[meta1['pert_iname'].isin(pert_inames_final)]
exp2final = meta2[meta2['pert_iname'].isin(pert_inames_final)]

# Retain only correct cell lines
cells1final = exp1final[exp1final['cell_id'].isin(['A549','HEPG2','WPMY','WPMY-1'])]
cells2final = exp2final[exp2final['cell_id'].isin(['A549','HEPG2','WPMY','WPMY-1'])]
```

```python
# Get sig ids
sig_ids = cells1final['sig_id'].tolist() + cells2final['sig_id'].tolist()
sig_ids = list(set(sig_ids))
sig_ids
```

```python
# Save sig ids
def write_to_csv(sig_ids):
    with open('../Output/sig_ids.csv', 'w') as csvfile:
        for i_d in sig_ids:
            csvfile.write(i_d + '\n')

write_to_csv(sig_ids)
```

```python
# Save file for mapping from sig id to experiment
cells1final.to_csv("../Output/sig_ids_overlap_1.csv")
cells2final.to_csv("../Output/sig_ids_overlap_2.csv")
```
