```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
```

## Using Tox21 only
-------------------------------------
|     Active     |    Inactive      |
|  ------------  |  --------------  |
|        917     |        5159      |
-------------------------------------

```python
df = pd.read_csv('cv-results.csv', index_col=0)
df = df[df['scoring'].isin(['test_accuracy', 'test_balanced_accuracy', 'test_recall', 'test_precision', 'test_roc_auc'])]
```

```python
grid = sns.FacetGrid(df, col="method", hue='scoring', col_wrap=4, height=5)
grid.map(sns.lineplot, 'n', 'value')
plt.legend()
```
